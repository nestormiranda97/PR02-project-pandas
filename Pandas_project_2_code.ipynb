# Let's start cleaning data

### I will first remove the 'Case Number.2' column if it's a duplicate from 'Case Number.1' 

```python
df["exists"] = df.drop("Case Number.2", 1).isin(df["Case Number.2"]).any(1)
# Drop the Case number column to have a dataframe without this column and with isin test if the value of Case Number.2
# exists in this new dataframe
# We can see that the data is true for all columns meaning that even though we drop the column Case.Number.2
# when we inspect the df, we find the same values, meaning that they are repeated

df.groupby('exists').size()

#Case Number.2 repeats for all values of Case Number1
```

```python
# Let's drop 'Case Number.2' that we proved irrelevant and the new column 'exists' that we created to check if the value was repeated
df = df.drop(['exists','Case Number.2'], axis=1)
```

```python
# I suspect that 'Unnamed 22 and Unnamed 23 are also irrelevant but they have 1 row that is not null,
# I want to see what they contain
df[df['Unnamed: 22'].notnull()]
```

```python
df[df['Unnamed: 23'].notnull()]

## From what I can see they do not contain any relevant information therefore I'm deleting them
```

```python
# I'm dropping 'Unnamed: 22' and 'Unnamed: 23' columns because they are useless
df = df.drop(['Unnamed: 22','Unnamed: 23'], axis=1)
```

```python
# I inspect the 'Time' column to see what's on it but I will keep it
df['Time']
```

## Now our df is more clean

```python
## Let's get some insights
df['Country'].value_counts()

# USA is the country with the most attacks
#There are many countries, some with little attacks, we should try to merge those small countries
```

```python
set(df['Country'])
# There are too many countries
```

```python
# I am grouping together in an "Other" category the countries that are present in less than 1.5% of the total count
threshold_percent = 1.5

series = pd.value_counts(df['Country'])
mask = (series / series.sum() * 100).lt(threshold_percent)
df = df.assign(country_grouped = np.where(df['Country'].isin(series[mask].index),'Other', df['Country']))
```

```python
df.country_grouped.value_counts()
```

```python
# I want to see how many of the attacks are fatal
df['Fatal (Y/N)'].value_counts()
```

```python
# We need to do some cleaning and combine the ocurrences where N has spaces or is in lowercase
```

```python
df['Fatal (Y/N)'].replace(" N", "N", inplace=True, )
df['Fatal (Y/N)'].replace("  N", "N", inplace=True, )
df['Fatal (Y/N)'].replace("n", "N", inplace=True, )
df['Fatal (Y/N)'].replace("N ", "N", inplace=True, )

df['Fatal (Y/N)'].value_counts()
```

```python
# I'm droping the rows where we don't have any info on Fatal(Y/N), null values
df.drop(df[df['Fatal (Y/N)'] == 'UNKNOWN'].index, inplace=True)
df.drop(df[df['Fatal (Y/N)'] == 'F'].index, inplace=True)
df.drop(df[df['Fatal (Y/N)'] == '#VALUE!'].index, inplace=True)
```
